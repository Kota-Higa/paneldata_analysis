```python
#必要パッケージの取り込み
import pandas as pd
import statsmodels.api as sm 

#データの読み込み
data = "./data/Economic Data - 9 Countries (1980-2020).csv"

df = pd.read_csv(data)

#表示(確認)
df
```

```python
#countryとyearを列の前に
df = df.set_index(['country', 'year'])

#表示(確認)
df
```

```python
#Pooled OLS
from linearmodels import PooledOLS
exog_vars=['unemploymentrate','percapitaincome'] 
mod=PooledOLS(df.gdppercent,sm.add_constant(df[exog_vars])) 
pooled_res = mod.fit()

#表示
print(pooled_res)
```

```python
#固定効果モデル
from linearmodels import PanelOLS
exog_vars=['unemploymentrate','percapitaincome'] 
mod=PanelOLS(df.gdppercent,sm.add_constant(df[exog_vars])) 
fe_res = mod.fit()

#表示
print(fe_res)
```

```python
#変動効果モデル
from linearmodels import RandomEffects
exog_vars=['unemploymentrate','percapitaincome'] 
mod=RandomEffects(df.gdppercent,sm.add_constant(df[exog_vars])) 
re_res = mod.fit()

#表示
print(re_res)
```

```python
#Pooled OLS（Pooled）,固定効果モデル(FE),変動効果モデル(RE)の比較
from linearmodels.panel import compare 

#表示
print(compare({'Pooled':pooled_res,'FE':fe_res,'RE':re_res}))
```
